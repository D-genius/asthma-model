## Chilhood Asthma in Kenya 

```python
import pandas as pd

# Load the dataset
file_path = 'assets/split_part_1.csv'
df = pd.read_csv(file_path)

# Inspect the data
print("DataFrame Head:")
print(df.head())
print("\nDataFrame Info:")
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import altair as alt

# --- 1. Data Selection and Preprocessing ---
# Select key variables
key_columns = ['Age', 'Gender', 'Allergies', 'Air_Pollution_Level', 'County', 'Has_Asthma']
df_model = df[key_columns].copy()

# Handle missing values in Allergies: assume NaN means 'None'
df_model['Allergies'] = df_model['Allergies'].fillna('None')

# One-hot encode categorical variables
categorical_cols = ['Gender', 'Allergies', 'Air_Pollution_Level', 'County']
df_encoded = pd.get_dummies(df_model, columns=categorical_cols, drop_first=True)

# Define features (X) and target (y)
X = df_encoded.drop('Has_Asthma', axis=1)
y = df_encoded['Has_Asthma']

# --- 2. Split Data ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print("--- Data Split ---")
print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}\n")

# --- 3. Train Random Forest Classifier ---
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
print("--- Model Training Complete (Random Forest Classifier) ---\n")

# --- 4. Model Evaluation ---
y_pred = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)

print("--- Model Evaluation Metrics (Random Forest Classifier) ---")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# --- 5. Predict Risk Score for each County ---
# Predict probabilities for the entire dataset
full_pred_proba = rf_model.predict_proba(X)[:, 1]

# Add probabilities and original County back to the full dataset for aggregation
df_model['Predicted_Risk'] = full_pred_proba

# Calculate the mean predicted risk (probability of Has_Asthma=1) for each County
county_risk_scores = df_model.groupby('County')['Predicted_Risk'].mean().reset_index()
county_risk_scores.columns = ['County', 'Mean_Predicted_Risk']

# Sort the counties by risk for better visualization
county_risk_scores = county_risk_scores.sort_values(by='Mean_Predicted_Risk', ascending=False)

# Save the county risk scores for the user's Choropleth map
county_risk_scores.to_csv('county_risk_scores.csv', index=False)
print("\nCounty risk scores saved to county_risk_scores.csv for Choropleth visualization.")
```

```python
# --- 6. Visualization 1: Bar Chart of Mean Predicted Risk per County (Alternative to Choropleth) ---

base_risk = alt.Chart(county_risk_scores).encode(
    x=alt.X('Mean_Predicted_Risk', title='Mean Predicted Risk (Prob. of Asthma)'),
    y=alt.Y('County', title='County', sort='-x'),
    tooltip=['County', alt.Tooltip('Mean_Predicted_Risk', format='.3f')]
).properties(
    title='Model Predicted Mean Asthma Risk by County of children Kenya'
)

bar_chart = base_risk.mark_bar().encode(
    color=alt.Color('Mean_Predicted_Risk', scale=alt.Scale(range='heatmap'), legend=alt.Legend(title="Risk Score"))
)

text = base_risk.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudge text to the right
).encode(
    text=alt.Text('Mean_Predicted_Risk', format='.2f'),
    color=alt.value('black')
)

chart_risk = (bar_chart + text).properties(height=700)
chart_risk.show()
chart_risk.save('county_risk_bar_chart.json')
print("Bar chart of county risk saved to county_risk_bar_chart.json.")

# --- 7. Visualization 2: Distribution Chart (Histogram) of Predicted Probabilities ---
risk_distribution_df = pd.DataFrame({'Predicted_Probability': full_pred_proba})
distribution_chart = alt.Chart(risk_distribution_df).mark_bar().encode(
    alt.X("Predicted_Probability:Q", bin=alt.Bin(maxbins=50), title='Predicted Asthma Probability'),
    alt.Y("count()", title='Count of Individuals'),
    tooltip=['Predicted_Probability', 'count()']
).properties(
    title='Distribution of Predicted Asthma Risk Probabilities of Children in Kenya'
)

distribution_chart.show()
distribution_chart.save('predicted_risk_distribution.json')
print("Distribution chart of predicted probabilities saved to predicted_risk_distribution.json.")
```
